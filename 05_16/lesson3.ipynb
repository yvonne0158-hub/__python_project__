```python
#學生總分300
#有些學生可加5%
#如果輸入分數無法轉成整數
while True:
    try:
        scores = int(input("請輸入學生分數(最高300分):"))
        break
    except ValueError:
        print("輸入錯誤，請輸入整數")
       
is_add = input("學生是否符合加分條件?(y,n)")

if  is_add == "y": #單向選擇
    scores = int(scores * 1.05 + 0.5)

print("學生分數", scores)

#超過300分，上限取300
if scores > 300:
    scores = 300

print("學生分數", scores)
```
